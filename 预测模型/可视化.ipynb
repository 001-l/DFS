```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#  AUC 
models = ['Random Forest', 'Gradient Boosting Trees', 'LightGBM', 'CatBoost', 'XGBoost']
auc_no_semantic = [0.7124, 0.7083, 0.7177, 0.6671, 0.6977]
auc_with_semantic = [0.7394, 0.7308, 0.7345, 0.7422, 0.7234]
relative_gain = [3.79, 3.18, 2.34, 11.26, 3.68]

# 
title_fontsize = 14
label_fontsize = 12
tick_fontsize = 10
value_fontsize = 10

# 
fig, axs = plt.subplots(1, 2, figsize=(14, 6))  # 

# ========== AUC  ==========
x = np.arange(len(models))
width = 0.35
bars1 = axs[0].bar(x - width/2, auc_no_semantic, width, label='Without Semantic', color='#A6CEE3')
bars2 = axs[0].bar(x + width/2, auc_with_semantic, width, label='With Semantic', color='#1F78B4')

# 
for bar in bars1 + bars2:
    height = bar.get_height()
    axs[0].text(bar.get_x() + bar.get_width()/2, height + 0.002, f'{height:.3f}', 
                ha='center', va='bottom', fontsize=value_fontsize)

axs[0].set_ylabel('AUC Score', fontsize=label_fontsize)
axs[0].set_title('AUC Performance Before and After Semantic Enhancement', fontsize=title_fontsize)
axs[0].set_xticks(x)
axs[0].set_xticklabels(models, fontsize=tick_fontsize)
axs[0].set_ylim(0.00, 1.00)
axs[0].legend(fontsize=10)
axs[0].grid(axis='y', linestyle='--', alpha=0.5)

# ==========  ==========
bars = axs[1].bar(x, relative_gain, width=0.6, color='#1F78B4')

for bar in bars:
    height = bar.get_height()
    axs[1].text(bar.get_x() + bar.get_width()/2, height + 0.3, f'{height:.2f}%', 
                ha='center', va='bottom', fontsize=value_fontsize)

axs[1].set_ylabel('Relative Gain (%)', fontsize=label_fontsize)
axs[1].set_title('Relative AUC Gain with Semantic Enhancement', fontsize=title_fontsize)
axs[1].set_xticks(x)
axs[1].set_xticklabels(models, fontsize=tick_fontsize)
axs[1].set_ylim(0, max(relative_gain) + 3)
axs[1].grid(axis='y', linestyle='--', alpha=0.5)

# 
plt.tight_layout()
plt.savefig("_AUC_.png", dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import auc

# 
models = {
    'rf': 'Random Forest',
    'gbt': 'Gradient Boosting Trees',
    'lgbm': 'LightGBM',
    'catb': 'CatBoost',
    'xgb': 'XGBoost'
}

colors = {
    'rf': 'tab:blue',
    'gbt': 'tab:orange',
    'lgbm': 'tab:green',
    'catb': 'tab:red',
    'xgb': 'tab:purple'
}

# 
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

# 
for i, (model_key, model_name) in enumerate(models.items()):
    ax = axes[i]
    
    # 
    df_plain = pd.read_csv(f'{model_key}_bjd_roc_curve.csv')
    auc_plain = auc(df_plain['FPR'], df_plain['TPR'])
    ax.plot(df_plain['FPR'], df_plain['TPR'], linestyle='-', color=colors[model_key],
             label=f'Before semantic (AUC={auc_plain:.3f})')

    # 
    df_semantic = pd.read_csv(f'{model_key}_yy_roc_curve.csv')
    auc_semantic = auc(df_semantic['FPR'], df_semantic['TPR'])
    ax.plot(df_semantic['FPR'], df_semantic['TPR'], linestyle='--', color=colors[model_key],
             label=f'After semantic (AUC={auc_semantic:.3f})')

    # 
    ax.plot([0, 1], [0, 1], linestyle='--', color='gray')
    
    ax.set_title(model_name, fontsize=12)
    ax.set_xlabel('FPR')
    ax.set_ylabel('TPR')
    ax.legend(fontsize=9)
    ax.grid(True)

# 6
if len(models) < len(axes):
    for j in range(len(models), len(axes)):
        fig.delaxes(axes[j])

# 
plt.suptitle('ROC Curves With and Without Semantic Enhancement', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig("_ROC_.png", dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import auc

# 
models = {
    'rf': 'Random Forest',
    'gbt': 'Gradient Boosting Trees',
    'lgbm': 'LightGBM',
    'catb': 'CatBoost',
    'xgb': 'XGBoost'
}

colors = {
    'rf': 'tab:blue',
    'gbt': 'tab:orange',
    'lgbm': 'tab:green',
    'catb': 'tab:red',
    'xgb': 'tab:purple'
}

for model_key, model_name in models.items():
    plt.figure(figsize=(6, 5))
    
    # 
    df_plain = pd.read_csv(f'{model_key}_bjd_roc_curve.csv')
    auc_plain = auc(df_plain['FPR'], df_plain['TPR'])
    plt.plot(df_plain['FPR'], df_plain['TPR'], linestyle='-', color=colors[model_key],
             label=f'Before Semantic (AUC={auc_plain:.3f})')

    # 
    df_semantic = pd.read_csv(f'{model_key}_yy_roc_curve.csv')
    auc_semantic = auc(df_semantic['FPR'], df_semantic['TPR'])
    plt.plot(df_semantic['FPR'], df_semantic['TPR'], linestyle='--', color=colors[model_key],
             label=f'After Semantic (AUC={auc_semantic:.3f})')

    # 
    plt.plot([0, 1], [0, 1], linestyle='--', color='gray')

    # 
    plt.xlabel('False Positive Rate (FPR)', fontsize=16)
    plt.ylabel('True Positive Rate (TPR)', fontsize=16)
    plt.title(f'ROC Curve: {model_name}', fontsize=16)
    plt.legend(loc='lower right', fontsize=13)
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)    
    plt.grid(True)
    plt.tight_layout()

    # 
    filename = f"ROC_{model_key}_Semantic Augmentation.svg"
    #plt.savefig(filename, dpi=300)
    plt.savefig(filename, format='svg')
    plt.close()  # 
```
